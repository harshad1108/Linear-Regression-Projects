```python
# Multiple linear regression

X_m=df[['Total Ad Spend','Order Count','Return Amount']]
y=df['Total Sales']

X_m_train,X_m_test,y_train,y_test=train_test_split(X_m,y,test_size=0.2,random_state=42)

model_m=LinearRegression()
model_m.fit(X_m_train,y_train)

y_pred_m=model_m.predict(X_m_test)

mse=mean_squared_error(y_test,y_pred_m)
mae=mean_absolute_error(y_test,y_pred_m)
rmse=np.sqrt(mse)
r2=r2_score(y_test,y_pred_m)

print("Slope:",model_m.coef_)
print("Intercept:",model_m.intercept_)
print("MSE:",mse)
print("MAE:",mae)
print("RMSE:",rmse)
print("R2:",r2)
```
